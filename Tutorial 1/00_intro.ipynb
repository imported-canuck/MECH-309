### Set up Python on your machine (short reminder)

Below are very brief, high-level hints for setting up a Python environment.  
You do **not** need a perfect setup—just something that lets you run Python scripts and install packages.

#### Linux
Most Linux distributions already come with Python installed.  
You can usually verify this by running:
```bash
python3 --version
```
If needed, install Python and pip using your package manager (e.g. apt, dnf, pacman).

#### macOS
The recommended approach is to install Python via a package manager such as Homebrew.
Once Homebrew is installed, Python can be installed with:
```bash
brew install python
```

#### Windows
I do not actively use Windows.
If you have not set up Python yet, the easiest and most robust solution is to install Windows Subsystem for Linux (WSL) and follow the Linux instructions above :D


### Package environment
One of the biggest advantages of Python is its rich and well-maintained ecosystem of packages and libraries.  
The main downside is that managing these packages can sometimes become cumbersome, especially when different projects require different versions.

Python comes with its own package manager, **`pip`**, which works well—particularly when used together with **virtual environments**.

Another option, which I personally recommend, is **Miniforge3**. It provides a lightweight Conda-based environment similar to Anaconda, but without unnecessary extras, and it uses the **conda-forge** channel by default. 
This often results in more reliable and consistent package installations, especially for scientific computing.


### Required Python packages
For this course, we will mainly rely on the following core scientific Python packages:

- **NumPy** – efficient array operations, linear algebra, and numerical computations  
- **SciPy** – numerical algorithms such as solvers, optimization, integration, and interpolation  
- **Matplotlib** – plotting and visualization of results  

Together, these packages form the foundation of the scientific Python ecosystem and are sufficient for all assignments and projects in this course.


### `.py` vs `.ipynb`

Python code can be written either as plain script files (`.py`) or as Jupyter notebooks (`.ipynb`).  
Both formats are useful, but they serve different purposes.

**`.py` files**
- Best for clean, reproducible code
- Easy to version control (Git)
- Suitable for larger projects and scripts
- Less interactive for exploration

**`.ipynb` files (Jupyter notebooks)**
- Excellent for exploration, learning, and visualization
- Combine code, text, equations, and plots in one place
- Very convenient for tutorials and reports
- Harder to version control cleanly
- Execution order can be confusing if cells are run out of order
- Additional packages for `.ipynb` files: `jupyter` and `ipykernel`

**Recommendation:**  
Use **Jupyter notebooks** for learning, experimentation, and visualization, and switch to **`.py` files** once the code becomes larger or more structured.
